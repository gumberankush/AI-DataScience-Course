# Simulating coin flip flop and plotting

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
output = ['1', '0']
tosses = ''.join([output[i] for i in np.random.randint(2, size=10000)])
tosses
```

```python
res = []
for i in range(1, 10001):
    output = np.random.randint(0, 2, i)
    prob = output.sum()/i
    res.append(prob)
res
```

```python
fig = plt.figure(figsize=(20,5))
plt.plot(res)
plt.show()
```
