# K-Means Clustering

## Import Libraries and Dataset

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import preprocessing
from sklearn.cluster import KMeans

import warnings

warnings.filterwarnings('ignore')
#importing the dataset

dataset = pd.read_csv('/Users/gumber/Documents/Personal_Repository/AI-DataScience-Course/ML/data/Cust_Spend_Data.csv')
#creating a duplicate dataset to work on
dataset1 = dataset
# peeking at the dataset
dataset1.head()
```

```python
#Descriptive stats of the variables in data
dataset1.describe()
```

## Feature Scaling

```python
# As name and cus id are not improtant. So scaling the remaining data
data = dataset1.iloc[:, 2:6].values

dataset1_standardzied = preprocessing.scale(data)
dataset1_standardzied = pd.DataFrame(dataset1_standardzied)
print(dataset1_standardzied)
```

## WSS Plot or Elbow Curve for Optimal cluster identification

```python
# find the appropriate number of clusters
plt.figure(figsize=(10, 8))

# within sum of square
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(dataset1_standardzied)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WIthin Sum of Square')
plt.show()
```

## K-Means Clustering & Cluster Information

```python
# The optimal number of cluster should be 3 as the elbow is being created at 3

kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(dataset1_standardzied)
y_kmeans

# Adding cluster to the original dataframe
cluster = pd.DataFrame(y_kmeans)

dataset1['cluster'] = cluster
dataset1.head()

# Mean of cluster
dataset1.groupby('cluster').mean()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(dataset1_standardzied)
y_kmeans

# Adding cluster to the original dataframe
cluster = pd.DataFrame(y_kmeans)

dataset1['cluster'] = cluster
dataset1.head()

# Mean of cluster
dataset1.groupby('cluster').mean()
```
